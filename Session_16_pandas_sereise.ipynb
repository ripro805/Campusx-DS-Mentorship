### What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,

### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# string
country = ['India', 'USA', 'Japan', 'Russia']
pd.Series(country)
```

```python
# integers
runs=[13,24,25,60,24]
runs_ser=pd.Series(runs)
runs_ser
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name


marks=pd.Series(marks,index=subjects,name="Ripro's Marks")
marks
```

### Series from dict

```python
marks={
    "maths":67,
    "english":57,
    "science":89,
    "hindi":100
}
pd.Series(marks,name="Ripro's Marks")
```
